WELCOME TO FILTERS IN PANDAS

```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(people)
```

```python
filt=(df['last']=='Doe')
```

```python
df[filt]
```

```python
df.loc[filt]
```

```python
df.loc[filt,'last']
```

```python
df.loc[filt,['first','email']]
```
