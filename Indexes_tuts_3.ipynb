WELCOME TO INDEX

```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(people)
```

```python
df
```

```python
df['email']
```

```python
df[['email','last']]
```

```python
df.set_index("email")
```

```python
df
```

```python
#for permanently changing the index
df.set_index("email",inplace=True)
```

```python
df
```

```python
df.index
```

```python
df.loc["JaneDoe@email.com",'last']
```

```python
df.loc[["JaneDoe@email.com","JohnDoe@email.com"],['last','first']]
```

```python
df.iloc[0:7,0:4]
```

```python
df.iloc[0,-1]
```

```python
df.iloc[1,-1]
```

```python
df.reset_index(inplace=True)
```
